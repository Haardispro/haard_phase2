## Solve
- The review report mentions compilation parameters
- Use that to compile the given source with the offcial makefile from ChipWhisperer - `make PLATFORM=CWNANO CRYPTO_TARGET=NONE -j OPT=1`
- Reverse the binary to see `OPT=1` has made the constant-time password comparison to non-constant
- There is no early breakout but comparison of each character can be noticed alone by the horizonatal segmentation of the 100 traces and vertical porfiling of each window as shown

```python
import numpy as np
import string
from tqdm import tqdm
import matplotlib.pyplot as plt
import os

traces_dir = "false_promise_traces"
chrset = string.printable

print(f"Loading {len(chrset)} traces from {traces_dir}/")
all_traces = {}

for c in tqdm(chrset, desc="Loading traces"):
    repeated_char = (c * 20).encode()
    filename = f"passwd_sent-{repeated_char.hex()}.npy"
    filepath = os.path.join(traces_dir, filename)
    
    if os.path.exists(filepath):
        trace = np.load(filepath)
        all_traces[c] = trace
    else:
        print(f"Warning: File not found for char '{c}': {filepath}")

print(f"Loaded {len(all_traces)} traces")

ref_trace_full = all_traces['7']
ref_pattern = ref_trace_full[27:40]
pattern_size = len(ref_pattern)

print(f"Reference pattern length: {pattern_size}")

def find_match_windows(trace, pattern, start_sample=13, target_matches=20, sad_threshold=10):
    pattern_len = len(pattern)
    raw_matches = []
    
    for i in range(start_sample, len(trace) - pattern_len + 1):
        window = trace[i:i + pattern_len]
        sad = np.sum(np.abs(pattern - window))
        raw_matches.append((i, i + pattern_len, sad))
    
    raw_matches.sort(key=lambda x: x[2])
    
    matches = []
    MIN_MATCH_SPACING = 5
    
    for start, end, sad in raw_matches:
        too_close = False
        for existing_start, existing_end, _ in matches:
            if abs(start - existing_start) < MIN_MATCH_SPACING:
                too_close = True
                break
        
        if not too_close:
            matches.append((start, end, sad))
            if len(matches) == target_matches:
                break
    
    matches.sort(key=lambda x: x[0])
    return matches

print("\nFinding match windows for all characters...")
trace_windows = {}
for c in tqdm(chrset, desc="Finding windows"):
    windows = find_match_windows(all_traces[c], ref_pattern)
    trace_windows[c] = windows

example_chars = ['7', 'a']
fig, axes = plt.subplots(len(example_chars), 1, figsize=(14, 6))

for idx, c in enumerate(example_chars):
    if c not in all_traces:
        continue
    
    ax = axes[idx]
    trace = all_traces[c]
    windows = trace_windows[c]
    print(windows)
    
    ax.plot(trace, 'b-', alpha=0.6, linewidth=0.8, label='Trace')
    
    for i, (start, end, sad) in enumerate(windows):
        ax.axvspan(start, end, alpha=0.2, color='red')
        ax.plot([start + (end-start)//2], [trace[start + (end-start)//2]], 
                'ro', markersize=4)
    
    ax.set_title(f"Character: '{c}' - {len(windows)} windows detected")
    ax.set_xlabel("Sample Index")
    ax.set_ylabel("Amplitude")
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("window_detection_examples.png", dpi=150)
print("Saved: window_detection_examples.png")
plt.show()

print("\nPerforming password recovery...")
guessed_pass = ""
position_analysis = []

for pos in range(20):
    position_sads = {}
    
    for c in chrset:
        windows = trace_windows[c]
        if pos < len(windows):
            start, end, sad = windows[pos]
            position_sads[c] = sad
    
    if position_sads:
        best_char = max(position_sads.items(), key=lambda x: x[1])
        guessed_pass += best_char[0]
        position_analysis.append({
            'pos': pos,
            'char': best_char[0],
            'sad': best_char[1],
            'all_sads': position_sads.copy()
        })
        print(f"Position {pos:2d}: '{best_char[0]}' (SAD = {best_char[1]:.2f})")

print(f"\n{'='*60}")
print(f"Final recovered password: {guessed_pass}")
print(f"As bytes: {guessed_pass.encode()}")
print(f"{'='*60}")

print("\nGenerating vertical profiling visualization...")

fig, axes = plt.subplots(4, 5, figsize=(24, 16))
axes = axes.flatten()

for pos in range(20):
    ax = axes[pos]
    pos_data = position_analysis[pos]
    all_sads = pos_data['all_sads']
    
    chars_ordered = [c for c in chrset if c in all_sads]
    sads_ordered = [all_sads[c] for c in chars_ordered]
    
    bars = ax.bar(range(len(chars_ordered)), sads_ordered, color='lightblue', edgecolor='gray', linewidth=0.5)
    
    selected_char = pos_data['char']
    if selected_char in chars_ordered:
        selected_idx = chars_ordered.index(selected_char)
        bars[selected_idx].set_color('lime')
        bars[selected_idx].set_edgecolor('darkgreen')
        bars[selected_idx].set_linewidth(2.5)
    
    ax.set_title(f"Position {pos}: '{pos_data['char']}' (SAD={pos_data['sad']:.1f})", 
                 fontsize=11, fontweight='bold')
    ax.set_xlabel("Characters (ASCII order)", fontsize=9)
    ax.set_ylabel("SAD Value", fontsize=9)
    ax.grid(True, alpha=0.3, axis='y')
    
    tick_positions = list(range(0, len(chars_ordered), 10))
    ax.set_xticks(tick_positions)
    tick_labels = []
    for i in tick_positions:
        c = chars_ordered[i]
        if c in ['\t', '\n', '\r', '\v', '\f']:
            tick_labels.append(repr(c))
        else:
            tick_labels.append(c)
    ax.set_xticklabels(tick_labels, fontsize=7, rotation=45)

plt.suptitle("Vertical Profiling: SAD Values for ALL Characters (ASCII Order) at Each Password Position", 
             fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig("vertical_profiling_ascii_order.png", dpi=150)
print("Saved: vertical_profiling_ascii_order.png")
plt.show()

print("\n✓ Analysis complete!")
```
